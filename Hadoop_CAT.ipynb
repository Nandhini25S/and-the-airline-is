```
csv_file_path = "/content/satisfaction.csv"
from pyspark.sql.types import IntegerType
df = spark.read.csv(csv_file_path,header=True)

df = df.withColumn("Seat comfort", df["Seat comfort"].cast(IntegerType()))
df = df.withColumn("Departure/Arrival time convenient", df["Departure/Arrival time convenient"].cast(IntegerType()))
df = df.withColumn("Food and drink", df["Food and drink"].cast(IntegerType()))
df = df.withColumn("Gate location", df["Gate location"].cast(IntegerType()))
df = df.withColumn("Inflight wifi service", df["Inflight wifi service"].cast(IntegerType()))
df = df.withColumn("Inflight entertainment", df["Inflight entertainment"].cast(IntegerType()))
df = df.withColumn("Online support", df["Online support"].cast(IntegerType()))
df = df.withColumn("Ease of Online booking", df["Ease of Online booking"].cast(IntegerType()))
df = df.withColumn("On-board service", df["On-board service"].cast(IntegerType()))
df = df.withColumn("Leg room service", df["Leg room service"].cast(IntegerType()))
df = df.withColumn("Baggage handling", df["Baggage handling"].cast(IntegerType()))
df = df.withColumn("Checkin service", df["Checkin service"].cast(IntegerType()))
df = df.withColumn("Cleanliness", df["Cleanliness"].cast(IntegerType()))
df = df.withColumn("Online boarding", df["Online boarding"].cast(IntegerType()))
df = df.withColumn("Departure Delay in Minutes", df["Departure Delay in Minutes"].cast(IntegerType()))
df = df.withColumn("Arrival Delay in Minutes", df["Arrival Delay in Minutes"].cast(IntegerType()))

df.printSchema()
df.count()
df.cache()
df.toPandas()
```

```
mean_arr_del = df.groupBy().avg('Arrival Delay in Minutes').take(1)[0][0]
mean_dep_del = df.groupBy().avg('Departure Delay in Minutes').take(1)[0][0]
print('Average departure delay (in minutes):', mean_dep_del)
print('Average arrival delay (in minutes): ', mean_arr_del)
```

```
feat = df.drop('id')
feat = feat.drop('label','Departure Delay in Minutes','Arrival Delay in Minutes', 'Flight Distance','Age','Customer Type','Type of Travel','Class','Gender')
from pyspark.sql.functions import col
from pyspark.sql.functions import when
feat = feat.withColumn("satisfaction_v2", when(df.satisfaction_v2 == "satisfied", 1).otherwise(0))
feat.show(3)
#feat.printSchema()
```

```
from pyspark.mllib.feature import StandardScaler
col_names = feat.columns
feat_rdd = feat.rdd.map(lambda row: row[0:])
scaler = StandardScaler(True,True).fit(feat_rdd)
scaled_features=scaler.transform(feat_rdd)
print(scaled_features.take(10))
```

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
feat_cols = ['Seat comfort','Departure/Arrival time convenient','Food and drink','Gate location','Inflight wifi service','Inflight entertainment','Online support','Ease of Online booking','On-board service','Leg room service','Baggage handling','Checkin service','Cleanliness','Online boarding']
vec_assembler = VectorAssembler(inputCols = feat_cols,outputCol='features_unscaled')
final = vec_assembler.transform(feat)
final.show()
final.printSchema()
```

```
train_final, test_final = final.randomSplit([0.70, 0.30], seed=15000)
train_final.show(3)
test_final.show(3)
print("Training Dataset Count: " + str(train_final.count()))
print("Test Dataset Count: " + str(test_final.count()))
```

```
#Linear_Regression
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(featuresCol = 'features_unscaled', labelCol='satisfaction_v2', maxIter=10, regParam=0.3, elasticNetParam=0.8)
lr_model = lr.fit(train_final)
print("Coefficients: " + str(lr_model.coefficients))
print("Intercept: " + str(lr_model.intercept))

trainingSummary = lr_model.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
print('Accuracy:', trainingSummary.r2*100)
```

```
#Decision_Tree_Regression
from pyspark.ml.regression import DecisionTreeRegressor
from pyspark.ml.evaluation import RegressionEvaluator

dt = DecisionTreeRegressor(featuresCol ='features_unscaled', labelCol = 'satisfaction_v2')
dt_model = dt.fit(train_final)
dt_predictions = dt_model.transform(test_final)
dt_evaluator = RegressionEvaluator(labelCol='satisfaction_v2', metricName="rmse")

rmse = dt_evaluator.evaluate(dt_predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)
```

```
#Logistic_Regression
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

lr = LogisticRegression(featuresCol = 'features_unscaled', labelCol = 'satisfaction_v2', maxIter=10)
lrModel = lr.fit(train_final)
lr_predictions = lrModel.transform(test_final)
lr_predictions.show(10)

multi_evaluator = MulticlassClassificationEvaluator(labelCol = 'satisfaction_v2', metricName = 'accuracy')
print('Logistic Regression Accuracy:', multi_evaluator.evaluate(lr_predictions))

import matplotlib.pyplot as plt
import numpy as np
beta = np.sort(lrModel.coefficients)
plt.plot(beta)
plt.ylabel('Beta Coefficients')
plt.show()
print('\n')
pr = trainingSummary.pr.toPandas()
plt.plot(pr['recall'],pr['precision'])
plt.ylabel('Precision')
plt.xlabel('Recall')
plt.show()
print('\n')

trainingSummary = lrModel.summary
roc = trainingSummary.roc.toPandas()
plt.plot(roc['FPR'],roc['TPR'])
plt.ylabel('False Positive Rate')
plt.xlabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
print('Training set areaUnderROC: ' + str(trainingSummary.areaUnderROC))
```

```
#Decision_tree_classifier
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

dt = DecisionTreeClassifier(featuresCol = 'features_unscaled', labelCol = 'satisfaction_v2', maxDepth = 3)
dtModel = dt.fit(train_final)
dt_predictions = dtModel.transform(test_final)
dt_predictions.show(3)

multi_evaluator = MulticlassClassificationEvaluator(labelCol = 'satisfaction_v2', metricName = 'accuracy')
print('Decision Tree Accuracy:', multi_evaluator.evaluate(dt_predictions))
```

```
#Random_forest_classifier
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
import matplotlib.pyplot as plt

rf = RandomForestClassifier(labelCol='satisfaction_v2', featuresCol='features_unscaled', maxDepth=3)
model = rf.fit(train_final)
rf_predictions = model.transform(test_final)
rf_predictions.show(3)

multi_evaluator = MulticlassClassificationEvaluator(labelCol = 'satisfaction_v2', metricName = 'accuracy')
print('Random Forest classifier Accuracy:', multi_evaluator.evaluate(rf_predictions))
```

```
#Gradient_boosted_tree
from pyspark.ml.classification import GBTClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

gb = GBTClassifier(labelCol = 'satisfaction_v2', featuresCol = 'features_unscaled')
gbModel = gb.fit(train_final)
gb_predictions = gbModel.transform(test_final)
gb_predictions.show(3)

multi_evaluator = MulticlassClassificationEvaluator(labelCol = 'satisfaction_v2', metricName = 'accuracy')
print('Gradient-boosted Trees Accuracy:', multi_evaluator.evaluate(gb_predictions))
```

```
#Naive_Bayes
from pyspark.ml.classification import NaiveBayes
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

nb = NaiveBayes(smoothing=1.0, modelType="multinomial")
model = nb.fit(train_final)
predictions = model.transform(test_final)
predictions.show()

evaluator = MulticlassClassificationEvaluator(labelCol='satisfaction_v2', metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Test set accuracy = " + str(accuracy))
```

```
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator

# Trains a k-means model.
kmeans = KMeans().setK(2).setSeed(1)
model = kmeans.fit(train_final)

# Make predictions
predictions = model.transform(train_final)

# Evaluate clustering by computing Silhouette score
evaluator = ClusteringEvaluator()

silhouette = evaluator.evaluate(predictions)
print("Silhouette with squared euclidean distance = " + str(silhouette))

# Shows the result.
centers = model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
    print(center)
```
